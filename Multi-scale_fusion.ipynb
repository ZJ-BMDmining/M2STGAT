```python
import numpy as np
import pandas as pd
```

```python
feature=pd.read_csv("finall_data_V04_200.csv",index_col=0,header=0)
feature.head()
```

```python
feature_colums=feature.columns
feature_colums
```

```python
matrix=pd.read_csv("Data_Correlation_V06_200.txt",index_col=None,header=0,sep="\t")
matrix.head()
```

```python
adj_matrix = pd.DataFrame(index=feature.columns, columns=feature.columns)
adj_matrix
```

```python
for _, row in matrix.iterrows():
    gene1 = row['row']
    gene2 = row['col']
    rho = row['rho']
    adj_matrix.loc[gene1, gene2] = rho
    adj_matrix.loc[gene2, gene1] = rho
```

```python
adj_matrix
```

```python
adj_matrix.values[np.diag_indices_from(adj_matrix)] = 1
adj_matrix
```

```python
adj_matrix = adj_matrix.fillna(0)
adj_matrix
```

```python
adj_matrix.to_csv("adj_MEGENA_V04_200.csv")
```

```python
sub_adj_matrix = adj_matrix.loc[genes, genes]
sub_adj_matrix
```
